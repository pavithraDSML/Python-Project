```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Campaign Performance:
```

```python
plt.figure(figsize=(10,6))
plt.scatter(Marketing_data['Acquisition_Cost'],Marketing_data['ROI'],linewidth=2.5,linestyles='--',marker='D',color='b')
plt.title('Acquisition_Cost vs ROI')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax=Marketing_data.groupby(['Campaign_Type','Channel_Used'])['Conversion_Rate'].mean().plot(kind='bar',color='yellow')
ax.bar_label(ax.containers[0])
plt.title('Conversion_Rate vs Channel_Used')
plt.xlabel('Conversion_Rate')
plt.ylabel('Channel_Used')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
Marketing_data.boxplot(column='Engagement_Score',by='Campaign_Type')
plt.xlabel('Campaign_Type')
plt.ylabel('Engagement_Score')
plt.color='red'
plt.show()
```

```python
ax=Marketing_data.groupby('Company')['ROI'].mean().plot(kind='bar',color='Orange')
ax.bar_label(ax.containers[0])
plt.title('Analyze the average ROI by Company')
plt.xlabel('Company')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
correlation_matrix=Marketing_data[['Engagement_Score','Conversion_Rate']].corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Customer Segmentation:
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=Marketing_data,x='Target_Audience',hue='Company')
plt.title(' distribution of Target_Audience.')
plt.xlabel('Target_Audience')
plt.ylabel('Company')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=Marketing_data,x='Language',y='Conversion_Rate',hue='Customer_Segment',estimator=max)
plt.title('highest Conversion_Rate for each Language user')
plt.xlabel('Language')
plt.ylabel('Conversion_Rate')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=Marketing_data,x='Acquisition_Cost',y='Customer_Segment',hue='Channel_Used')
plt.title('Customer_Segment categorized by Channel_Used')
plt.xlabel('Customer_Segment')
plt.ylabel('Engagement_Score')
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax=Marketing_data.groupby('Location')['Conversion_Rate'].mean().plot(kind='bar',color='maroon')
ax.bar_label(ax.containers[0])
plt.title('Average Conversion_Rate by Language')
plt.xlabel('Language')
plt.ylabel('Conversion_Rate')
plt.xticks(rotation=15)
plt.show()

```

```python
#Channel Effectiveness:
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=Marketing_data,x='Engagement_Score',y='Channel_Used',hue='Campaign_Type')
plt.title('Different Channels_Used segmented by Campaign_Type')
plt.xlabel('Engagement_Score')
plt.ylabel('Channel_Used')
plt.show()
```

```python
plt.figure(figsize=(8,6))
Marketing_data.groupby('Channel_Used')['ROI'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.title('Total ROI by Channel_Used')
plt.tight_layout()
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=Marketing_data,x='Clicks',y='Impressions',hue='Channel_Used')
plt.title('relationship between Clicks and Impressions')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.show()
```

```python
#Time-Based Analysis:
```

```python
sns.histplot(Marketing_data['Duration'],kde=True,color='green')
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(75,8))
sns.lineplot(data=Marketing_data, x='Date',  y='Conversion_Rate', marker='o', hue='Company', errorbar=None)
plt.title('Conversion_Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Conversion_Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(80,8))
plt.plot(Marketing_data['Date'],Marketing_data['Engagement_Score'],marker='v',color='red',markeredgecolor='k')
plt.title('The trend of Engagement_Score over Time')
plt.xlabel('Date')
plt.ylabel('Engagement_Score')
plt.xticks(rotation=90)
plt.show()
```

```python
#Geographic Analysis:
```

```python
plt.figure(figsize=(10,5))
ax=Marketing_data.groupby('Location')['Acquisition_Cost'].max().plot(kind='bar',color='green')
ax.bar_label(ax.containers[0])
plt.title(' Highest Conversion Rate ')
plt.xlabel('Location')
plt.ylabel('Conversion_Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=Marketing_data,x='Location',y='Conversion_Rate',hue='Target_Audience')
plt.title('Conversion_Rate by Location')
plt.xlabel('Location')
plt.ylabel('Conversion_Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
Marketing_data.groupby('Location')['ROI'].sum().plot(kind='pie',autopct='%1.2f%%')
plt.title('proportion of ROI by Location')
plt.tight_layout()
plt.ylabel('')
plt.show()
```
